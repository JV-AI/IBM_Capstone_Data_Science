```python
# Initialize the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Define helper functions for adding circles and markers
def add_circle(location, radius, color, popup_text):
    return folium.Circle(
        location=location,
        radius=radius,
        color=color,
        fill=True,
        fill_color=color
    ).add_child(folium.Popup(popup_text))

def add_marker(location, label, color):
    return folium.Marker(
        location=location,
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:{color};"><b>{label}</b></div>',
        )
    )

# Add circles for different launch sites
circle_ccafs_lc_40 = add_circle(ccafs_lc_40, 1000, '#FF5733', 'CCAFS LC-40')
circle_ccafs_slc_40 = add_circle(ccafs_slc_40, 1000, '#33FF57', 'CCAFS SLC-40')
circle_ksc_lc_39a = add_circle(ksc_lc_39a, 1000, '#3357FF', 'KSC LC-39A')
circle_vafb_slc_4e = add_circle(vafb_slc_4e, 1000, '#FF33A8', 'VAFB SLC-4E')
nasa_circle = add_circle(nasa_coordinate, 1000, '#FF8F33', 'NASA Johnson Space Center')

# Add markers for different launch sites
marker_ccafs_lc_40 = add_marker(ccafs_lc_40, 'CCAFS LC-40', '#d35400')
marker_ccafs_slc_40 = add_marker(ccafs_slc_40, 'CCAFS SLC-40', '#d35400')
marker_ksc_lc_39a = add_marker(ksc_lc_39a, 'KSC LC-39A', '#d35400')
marker_vafb_slc_4e = add_marker(vafb_slc_4e, 'VAFB SLC-4E', '#d35400')
nasa_marker = add_marker(nasa_coordinate, 'NASA JSC', '#d35400')

# Add circles and markers to the map
site_map.add_child(circle_ccafs_lc_40)
site_map.add_child(circle_ccafs_slc_40)
site_map.add_child(circle_ksc_lc_39a)
site_map.add_child(circle_vafb_slc_4e)
site_map.add_child(nasa_circle)

site_map.add_child(marker_ccafs_lc_40)
site_map.add_child(marker_ccafs_slc_40)
site_map.add_child(marker_ksc_lc_39a)
site_map.add_child(marker_vafb_slc_4e)
site_map.add_child(nasa_marker)

# Create marker clusters for successes and failures
success_cluster = MarkerCluster(name='Successes').add_to(site_map)
failure_cluster = MarkerCluster(name='Failures').add_to(site_map)

# Add dynamic markers based on the binary Class column in spacex_df
for _, row in spacex_df.iterrows():
    if row['class'] == 1:  # Success
        color = 'green'
        cluster = success_cluster
    else:  # Failure (Class == 0)
        color = 'red'
        cluster = failure_cluster

    # Add marker to the appropriate cluster
    folium.Marker(
        location=[row['Lat'], row['Long']],
        icon=folium.Icon(color=color),
        popup=f"Site: {row['Launch Site']}<br>Class: {'Success' if row['class'] == 1 else 'Failure'}"
    ).add_to(cluster)

# Add layer control to toggle clusters
folium.LayerControl().add_to(site_map)

# Display the map
site_map
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from folium.plugins import MousePosition

# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Add custom JavaScript for copy-paste functionality
copy_paste_js = """
<script>
    function copyToClipboard(content) {
        const tempInput = document.createElement("input");
        tempInput.style.position = "absolute";
        tempInput.style.left = "-1000px";
        tempInput.value = content;
        document.body.appendChild(tempInput);
        tempInput.select();
        document.execCommand("copy");
        document.body.removeChild(tempInput);
        alert("Copied to clipboard: " + content);
    }

    function addCopyFunctionality(map) {
        map.on("mousemove", function(e) {
            const lat = e.latlng.lat.toFixed(5);
            const lng = e.latlng.lng.toFixed(5);
            const coords = `Lat: ${lat}, Long: ${lng}`;
            map.on("click", function() {
                copyToClipboard(coords);
            });
        });
    }

    addCopyFunctionality(map);
</script>
"""

# Add JavaScript to the map
site_map.get_root().html.add_child(folium.Element(copy_paste_js))

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance
def calculate_distance(lat1, lon1, lat2, lon2):
    # Approximate radius of the Earth in kilometers
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordinates for the coastline and launch site
coastline_lat =  28.56405
coastline_lon = -80.56809
launch_site_lat = 28.56276
launch_site_lon = -80.57672

# Calculate the distance between the launch site and the coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Add a marker for the coastline and display the distance
coastline_coordinate = [coastline_lat, coastline_lon]

# Add a line connecting the launch site and the coastline
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], coastline_coordinate],
    color='blue',
    weight=2.5,
    opacity=1
).add_to(site_map)

# Add a marker at the coastline showing the distance
distance_marker = folium.Marker(
    location=coastline_coordinate,
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_coastline),
    )
)

# Add the distance marker to the map
site_map.add_child(distance_marker)

# Add a marker at the launch site
launch_site_marker = folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    icon=folium.Icon(color='blue', icon='rocket'),
    popup='Launch Site'
)
site_map.add_child(launch_site_marker)

# Display the map
site_map
```
